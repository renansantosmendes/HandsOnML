# Logistic Regression Classifier

 - In this notebook, an a classifier will be trained to the iris data set using Logistic Regression
 - This code can be found in book: "Hands-on Machine Learning with Scikit-Learn & TensorFlow"

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
```

### Importing the datasets

```python
iris = datasets.load_iris()
list(iris.keys())
```

### Spliting the dataset into a target and features

```python
X = iris['data'][:,3:]
y = (iris['target'] == 2).astype(np.int)
```

### Training the model

```python
log_reg = LogisticRegression()
log_reg.fit(X,y)
```

```python
X_new = np.linspace(0, 3, 1000).reshape(-1,1)
y_prob = log_reg.predict_proba(X_new)

plt.figure(figsize=(10,6))
plt.plot(X_new, y_prob[:,1], 'g-', label = 'Iris-Virginica')
plt.plot(X_new, y_prob[:,0], 'b--', label = 'Not Iris-Virginica')
plt.xlabel('Petal width (cm)', fontsize=20)
plt.ylabel('Probability', fontsize=20)
plt.legend(fontsize=18)
plt.show()
```

### Applying Softmax Regression

```python
X = iris['data'][:,(2,3)]
y = iris['target']

softmax_reg = LogisticRegression(multi_class='multinomial', solver = 'lbfgs', C = 10)
softmax_reg.fit(X,y)
```

#### Making predictions

```python
softmax_reg.predict([[5,2]])
```

```python
softmax_reg.predict_proba([[5,2]])
```
